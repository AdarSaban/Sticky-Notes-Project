```python
import pandas as pd
url = 'https://raw.githubusercontent.com/ShaiYona/Data-Science2021B/main/tirgulim/tirgul5/trees.csv'
treesData = pd.read_csv(url)
treesData.head()
```

```python
treesHeight = treesData['Height']
treesVolume = treesData['Volume']
```

```python
plt.plot(treesHeight,treesVolume,'bo') # 'bo' states blue circle
plt.title('Trees Volume vs Height')
plt.ylabel('Volume')
plt.xlabel('Height')
plt.show()
```

```python
# display the boxplot
myData = [treesData['Height'] , treesData['Volume']]
plt.boxplot(myData)
plt.ylim(-222,100)
a=2
# + is an outlier
# present 1st and 3rd quaters
# red line is the medline
```

```python
# histogram:
plt.hist(treesData['Height'] )
plt.ylabel('# num of appearances')
plt.xlabel('height')
plt.title('histogram of trees height')
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10*np.pi , 1000)
y = np.sin(x)
plt.plot(x,y)
```

```python
# we can plot several subplots
fig = plt.figure() # create a plot figure
# create the first of two panels and set current axis
import numpy as np
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))
plt.title('sin x')
# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), color = 'red' , linestyle = '--'); # --r will also work
plt.title('cos x',color = 'green', fontsize = 20) # note for color and fontsize
```

```python
# self excercise:
# plot y=x^2 from -10 to 10
# with x and y labels and title and legend
# with dashed yellow line
```

```python
x = np.linspace(-10,10,1000)
plt.plot(x,x**2,'--y')
plt.xlabel('x')
plt.ylabel('y=x^2')
plt.title('y=x^2')
```

```python
# we can control the axis limits:
plt.plot(x, np.sin(x))
plt.xlim(-4,4)
plt.ylim(-2,2)
# plt.axis('tight');
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
# adjust ratio
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```python
# we can add legend
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':r', label='cos(x)')
plt.legend();
```
