```python
import pandas as pd
url = 'https://raw.githubusercontent.com/ShaiYona/Data-Science2021B/main/tirgulim/tirgul5/trees.csv'
treesData = pd.read_csv(url)
treesData.head()
```

```python
treesHeight = treesData['Height']
treesVolume = treesData['Volume']
```

```python
fig = sns.scatterplot(treesHeight ,treesVolume ) # Label aquird from column titles
fig.set_title("Trees height vs volume")
```

```python
# display the boxplot
myData = treesData[['Height','Volume']]
ax = sns.boxplot(data = myData)

# + is an outlier
# present 1st and 3rd quaters
# ine is the medline
```

```python
sns.histplot(data = myData["Height"] , color='lime') # set width of bins
```

```python
sns.histplot(data = myData ) # set width of bins
```

```python
sns.histplot(data = myData ,  binwidth=5 , kde='true' ) # set width of bins
# kde - Kernel Density Estimation 
```

```python
# we could just plot the kde
sns.kdeplot(data = myData, fill = 'true')
```

```python
import numpy as np
import matplotlib.pyplot as plt
numHalfCycles = 10;
x = np.linspace(0,numHalfCycles*np.pi , 100*numHalfCycles+1)
y = np.sin(x)
fig = sns.lineplot(x,y)
fig.set_xlabel('x')
fig.set_ylabel('sin(x)')
fig.set_title("plot of sin(x)")
```

```python
numHalfCycles = 4;
x = np.linspace(0,numHalfCycles*np.pi , 100*numHalfCycles+1)
y = np.sin(x)
fig = sns.lineplot(x,y,color='r')
fig.set_xlabel('x')
fig.set_ylabel('sin(x)')
fig.set_title("plot with attributes")

fig.grid()
xticksValues = np.linspace(0,numHalfCycles*np.pi , (2*numHalfCycles)+1)
xticksNames = ['0','0.5pi','pi','1.5pi','2pi','2.5pi','3pi','3.5pi','4pi']
tickVals = plt.xticks(xticksValues,xticksNames,rotation = 45)
```

```python
# subplots:
fig, axes = plt.subplots(2, 2, figsize=(18, 10))
fig.suptitle('Using subplots')

xValues = np.linspace(0,numHalfCycles*np.pi , 100*numHalfCycles+1)
fig0 = sns.lineplot(ax = axes[0,0],x=xValues,y=np.sin(xValues)  )
fig0.set_title("first title example")

fig01 = sns.lineplot(ax = axes[0,1],x=xValues,y=np.exp(xValues), color = 'magenta'  )
fig01.set_title("third title example")
fig01.lines[0].set_linestyle("--")

fig10 = sns.lineplot(ax = axes[1,0],x=xValues,y=np.sin(xValues)  )
fig10 = sns.lineplot(ax = axes[1,0],x=xValues,y=np.cos(xValues) , color = 'lime'  )
fig10.set_title("second title example")
fig10.lines[1].set_linestyle("--")


fig11 = sns.lineplot(ax = axes[1,1],x=xValues,y=1.0/np.exp(xValues) , color = 'orange'  )
fig11.set_title("forth title example")
```

```python
# self excercise:
# plot y=x^2 from -10 to 10
# with x and y labels and title and legend
# with dashed yellow line
```

```python
hl = 10
ll = -10
numVals = 100*(hl-ll)+1
x = np.linspace(ll,hl , numVals)
y = np.square(x)
fig = sns.lineplot(x,y,color='y')
fig.lines[0].set_linestyle("--")
fig.set_title("y=x^2")
fig.set_xlabel('x')
fig.set_ylabel('x^2')
```
